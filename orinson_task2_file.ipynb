# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load Datasets

```python
df = pd.read_csv('https://raw.githubusercontent.com/Mhdthouyib123/code/refs/heads/main/random_weight_height_dataset%201.csv')
df
```

```python
df.head()
```

## Plotting a scatter plot to visualize the relationship between weight and height

```python
plt.figure(figsize=(6,6))
plt.scatter(df['Weight'] , df['Height'])
plt.xlabel('Weight of the person')
plt.ylabel('Height of the person')
```

```python
df.corr()
```

## Create pairwise plots for exploring relationships between variables

```python
import seaborn as sns
sns.pairplot(df)
```

```python
x = df[['Weight']]
x
```

```python
y = df['Height']
y
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train ,y_test = train_test_split(x , y , test_size=0.2 , random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(n_jobs=-1)
lr.fit(x_train , y_train)
```

```python
print("coefficient is = " , lr.coef_)
print("intercept is = " , lr.intercept_)
```

```python
plt.scatter(x_train , y_train)
plt.plot(x_train , lr.predict(x_train))
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
mse = mean_squared_error(y_test , y_pred)
mae = mean_absolute_error(y_test , y_pred)
rmse = np.sqrt(mse)
```

```python
print("MSE =" , mse)
print("MAE =" , mae)
print("RMSE = " , rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test  , y_pred)
score
```

```python
import statsmodels.api as sm
model = sm.OLS(y_train , x_train).fit()
prediction = model.predict(x_train)
prediction
```
